```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns

sentences = [
    'I love living by the Ocean shores',
    'Swimming is my favorite sport',
    'I love reading books a lot'
]

models = {
    'MiniLM': SentenceTransformer('all-MiniLM-L6-v2'),
    'MpNet': SentenceTransformer('all-mpnet-base-v2'),
    'BERT': SentenceTransformer('sentence-transformers/bert-base-nli-mean-tokens')
}

for name, model in models.items():
    embeddings = model.encode(sentences)
    sim_matrix = cosine_similarity(embeddings)

    print(f"\nCosine Similarity using {name}:\n", sim_matrix.round(2))

    # 🖼️ Visualize inside the loop
    plt.figure(figsize=(6, 4))
    sns.heatmap(sim_matrix, annot=True, xticklabels=sentences, yticklabels=sentences, cmap="Blues")
    plt.title(f"Cosine Similarity Matrix - {name}")
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```
